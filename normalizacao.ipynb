```python
import pandas as pd
import numpy as np

dfRed = pd.read_csv('dataset/winequality-red.csv', sep =';')
dfWhite = pd.read_csv('dataset/winequality-white.csv', sep =';')

display(dfRed)
display(dfWhite)
```

### Normalizando a base de dados

```python
def min_max_normalization(column):
    return (column - column.min()) / (column.max() - column.min())

# Aplica a normalização para todas as colunas exceto a quality
dfWhite_qual = dfWhite['quality']
dfWhite = dfWhite.drop(columns=['quality'])
dfWhite = dfWhite.apply(min_max_normalization)
dfWhite['quality'] = dfWhite_qual

dfRed_qual = dfRed['quality']
dfRed = dfRed.drop(columns=['quality'])
dfRed = dfRed.apply(min_max_normalization)
dfRed['quality'] = dfRed_qual

display(dfWhite)
display(dfRed)
```

Exporta a base de dados normalizada para csv

```python
dfRed.to_csv('dataset/dfRed.csv', sep=';', index=False)
dfWhite.to_csv('dataset/dfWhite.csv', sep=';', index=False)
```

Importa MLP e os métodos necessários

```python
from sklearn.neural_network import MLPClassifier # MLP
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

Prepara as bases de dados para treinamento

```python
# cria um dataframe sem a coluna qualidade
dfRed_temp = pd.DataFrame(dfRed,columns=dfRed.columns[:-1])
dfWhite_temp = pd.DataFrame(dfWhite,columns=dfWhite.columns[:-1])

#cria dataframe somente com a coluna qualidade
dfRed_qual = pd.DataFrame(dfRed,columns=['quality'])
dfWhite_qual = pd.DataFrame(dfWhite,columns=['quality'])
display(dfRed_temp.head())
display(dfWhite_temp.head())
```

Divide os dataframes em 50% teste, 25% treino e 25% validação

```python
# Dataframe Red Wine
# Dividir o conjunto de dados em 65% para treinamento e 35% para o resto
xRed_treino, xRed_resto, yRed_treino, yRed_resto = train_test_split(dfRed_temp,dfRed['quality'],test_size=0.35, stratify=dfRed['quality'])
# Divide o resto em 20% para teste e 15% para validação
xRed_validacao, xRed_teste, yRed_validacao, yRed_teste = train_test_split(xRed_resto, yRed_resto, test_size=0.42, stratify=yRed_resto, shuffle=True)
display(xRed_treino.shape)
display(xRed_teste.shape)
display(xRed_validacao.shape)

# Dataframe White Wine
# Divide o conjunto de dados em 65% para treinamento e 35% para o resto
xWhite_treino, xWhite_resto, yWhite_treino, yWhite_resto = train_test_split(dfWhite_temp,dfWhite['quality'],test_size=0.35, stratify=dfWhite['quality'])
# Divide o resto em 20% para teste e 15% para validação
xWhite_validacao, xWhite_teste, yWhite_validacao, yWhite_teste = train_test_split(xWhite_resto, yWhite_resto, test_size=0.42, stratify=yWhite_resto, shuffle=True)
display(xWhite_treino.shape)
display(xWhite_teste.shape)
display(xWhite_validacao.shape)
```

Prepara o GridSearchCV e busca os melhores parâmetros

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'hidden_layer_sizes': [(100,), (50, 50), (20,), (100, 50, 25)],
    'activation': ['relu', 'tanh', 'logistic', 'identity'],
    'solver': ['adam', 'sgd', 'lbfgs'],
    'learning_rate_init': [0.001, 0.01, 0.05, 0.1]
}

mlp_red = MLPClassifier(max_iter=200)
mlp_white = MLPClassifier(max_iter=200)

# Converte vetor de rótulos em matriz bidimensional
yRed_treino = np.array(yRed_treino).reshape(-1, 1)
grid_search_red = GridSearchCV(mlp_red, param_grid, cv=5)
grid_search_red.fit(xRed_treino, yRed_treino)

# Converte vetor de rótulos em matriz bidimensional
yWhite_treino = np.array(yWhite_treino).reshape(-1, 1)
yWhite_treino = list(yWhite_treino)
grid_search_white = GridSearchCV(mlp_white, param_grid, cv=5)
grid_search_white.fit(xWhite_treino, yWhite_treino)

```

```python
# Acessa os melhores resultados
best_params_red = grid_search_red.best_params_
best_model_red = grid_search_red.best_estimator_
print("Parametros vinho tinto:")
print(best_params_red)
print("Melhor acurácia no treino:")
print(grid_search_red.best_score_)

best_params_white = grid_search_white.best_params_
best_model_white = grid_search_white.best_estimator_
print("Parametros vinho branco:")
print(best_params_white)
print("Melhor acurácia no treino:")
print(grid_search_white.best_score_)


print("Acurácia melhor modelo Vinho tinto")
accuracy_test_red = best_model_red.score(xRed_teste, yRed_teste)
accuracy_validation_red = best_model_red.score(xRed_validacao, yRed_validacao)
print("TESTE:")
print(accuracy_test_red)
print("VALIDACAO:")
print(accuracy_validation_red)

print("Acurácia melhor modelo Vinho branco")
accuracy_test_white = best_model_white.score(xWhite_teste, yWhite_teste)
accuracy_validation_white = best_model_white.score(xWhite_validacao, yWhite_validacao)
print("TESTE:")
print(accuracy_test_white)
print("VALIDACAO:")
print(accuracy_validation_white)
```
