```python
import pandas as pd

dfRed = pd.read_csv('dataset/winequality-red.csv', sep =';')
dfWhite = pd.read_csv('dataset/winequality-white.csv', sep =';')

display(dfRed)
display(dfWhite)
```

### Normalizando a base de dados

```python
def min_max_normalization(column):
    return (column - column.min()) / (column.max() - column.min())

# Aplicar a normalização Min-Max em cada coluna do DataFrame
dfWhite = dfWhite.apply(min_max_normalization)
dfRed = dfRed.apply(min_max_normalization)

display(dfWhite)
display(dfRed)
```

Exporta a base de dados normalizada para csv

```python
dfWhite.to_csv('dataset/dfWhite.csv', sep=';', index=False)
dfRed.to_csv('dataset/dfRed.csv', sep=';', index=False)
```
